```python
%matplotlib inline

from matplotlib import pyplot as plt
import numpy as np
import os

## Store Results
source = '../Data/Results'
for root, dirnames, filenames in os.walk(source):
    trials = filenames

## Sort Trials:
trained_1chunk = sorted([x for x in trials if 'Trained' in x and '1chunk' in x], key=lambda x: x.split('cscore_')[1].split('.txt')[0], reverse=True )
trained_2chunk = sorted([x for x in trials if 'Trained' in x and '2chunk' in x], key=lambda x: x.split('cscore_')[1].split('.txt')[0], reverse=True )

untrained_1chunk = sorted([x for x in trials if 'Untrained' in x and '1chunk' in x], key=lambda x: x.split('cscore_')[1].split('.txt')[0], reverse=True )
untrained_2chunk = sorted([x for x in trials if 'Untrained' in x and '2chunk' in x], key=lambda x: x.split('cscore_')[1].split('.txt')[0], reverse=True )

## Borrowing heavily from: http://matplotlib.org/1.2.1/examples/api/barchart_demo.html
## For the grouped plots
```

# Untrained Results: 1 Chunk

```python
def plot_results( trial_set ):
    for trial_name in trial_set:
        
        trial = np.loadtxt( '../Data/Results/' + trial_name, delimiter = ',')
        means = np.mean( trial, axis = 0 )
        stds = np.std( trial, axis = 0 )
        
        accuracies = means[::2]
        acc_std = stds[::2]
        softcalls = means[1::2]
        sc_std = stds[1::2]
        N = 6
        
        ind = np.arange(N)
        width = 0.35
        
        fig = plt.figure()
        ax = fig.add_subplot(111)
        rects1 = ax.bar( ind, accuracies, width, color='r', yerr=acc_std )
        rects2 = ax.bar( ind+width, softcalls, width, color='g', yerr=sc_std )
        
        ax.set_ylabel( 'Value' )
        ax.set_title( trial_name.split('.txt')[0] )
        ax.set_xticks(ind+width)
        ax.set_xticklabels( ('First', 'Last', 'Random', 'HMM', 'Best', 'Ind' ) )
        
        ax.legend( (rects1[0], rects2[0]), ('Accuracy', 'Softcall') )
        
        def autolabel(rects):
                for rect in rects:
                    height = rect.get_height()
                    ax.text(rect.get_x()+rect.get_width()*1.0/2., 1.05*height, str(round(height,2)), ha='center', va='bottom')

        autolabel(rects1)
        autolabel(rects2)
        
        plt.show()
        
plot_results( untrained_1chunk )
    
```

# Untrained Results: 2 Chunk

```python
plot_results( untrained_2chunk )
```

# Trained 1Chunk

```python
plot_results( trained_1chunk )
```

# Trained 2Chunk

```python
plot_results( trained_2chunk )
```
