```python
%matplotlib inline

from matplotlib import pyplot as plt
import numpy as np
import os
import seaborn as sns

## Store Results
trials = []
source = '../Data/Results'
for root, dirnames, filenames in os.walk(source):
    trials.append( filenames )

trials = trials[0] # Select root files
## Sort Trials:
untrained_newset = sorted([x for x in trials if 'untrained' in x and 'newset' in x], key=lambda x: x.split('cscore_')[1].split('.txt')[0], reverse=True )
trained_new = sorted([x for x in trials if 'Trained' in x and 'New' in x], key=lambda x: x.split('cscore_')[1].split('.txt')[0], reverse=True )
```

# Line Plot Function

```python
def accuracy_by_filter_score( trial_set ):
    
    # Convert Groups of arrays into lists of scores by csore.
    first, last, random = [], [], []
    hmm, best, ind = [], [], []
    sample_sizes = []
    
    for trial_name in trial_set:
            
            trial = np.loadtxt( '../Data/Results/' + trial_name, delimiter = ',' )
            #accuracies = means[::2]
            #acc_std = stds[::2]
            #softcalls = means[1::2]
            #sc_std = stds[1::2]

            accuracies = trial[::2]
            softcalls = trial[1::2]
            
            f,l,r,h,b,i = accuracies
            first.append( f ); last.append( l ); random.append( r )
            hmm.append( h ) ; best.append( b ); ind.append( i )
         
            sample_sizes.append( trial_name.split('_')[3] )
            
    x = [ 0.9, 0.8, 0.7, 0.6, 0.5, 0.4, 0.3, 0.2, 0.1, 0.0 ]
    plt.plot( x, first, label='first' )
    plt.plot( x, last, label='last' ) 
    plt.plot( x, random, label='random' )
    
    plt.plot( x, hmm, label='hmm' )
    plt.plot( x, best, label='best' )
    plt.plot( x, ind, label='ind' )
    
    plt.title( trial_name.split('_')[0] + ' - ' + trial_name.split('_')[1] \
                + ' - ' + trial_name.split('_')[2], fontsize=14 )
   
    plt.xlabel( 'Chunk Score Cutoff', fontsize=14 )
    plt.ylabel( 'Accuracy', fontsize=14 )
    plt.ylim( [0,1.0] )
    plt.legend(loc='lower left', bbox_to_anchor=(1.02, 0), borderaxespad=0, fontsize=14)
    plt.gca().invert_xaxis()
    plt.show()

print untrained_newset
accuracy_by_filter_score( untrained_newset )

print trained_new
accuracy_by_filter_score( trained_new )
```

# Individual Trials by Cscore 

```python
def plot_results( trial_set ):
    ## Borrowing heavily from: http://matplotlib.org/1.2.1/examples/api/barchart_demo.html
    for trial_name in trial_set:
        
        trial = np.loadtxt( '../Data/Results/' + trial_name, delimiter = ',')
        #means = np.mean( trial, axis = 0 )
        #stds = np.std( trial, axis = 0 )
        
        #accuracies = means[::2]
        #acc_std = stds[::2]
        #softcalls = means[1::2]
        #sc_std = stds[1::2]
        
        accuracies = trial[::2]
        softcalls = trial[1::2]
        
        N = 6
        
        ind = np.arange(N)
        width = 0.35
        
        fig = plt.figure()
        ax = fig.add_subplot(111)
        rects1 = ax.bar( ind, accuracies, width, color='r' )#, yerr=acc_std )
        rects2 = ax.bar( ind+width, softcalls, width, color='g')#, yerr=sc_std )
        
        ax.set_ylabel( 'Value' )
        ax.set_title( trial_name.split('.txt')[0] )
        ax.set_xticks(ind+width)
        ax.set_xticklabels( ('First', 'Last', 'Random', 'HMM', 'Best', 'Ind' ) )
        plt.ylim( [0, 1.0] )
        
        ax.legend( (rects1[0], rects2[0]), ('Accuracy', 'Softcall'), loc='lower left', bbox_to_anchor=(1.02, 0), borderaxespad=0 )
        
        def autolabel(rects):
                for rect in rects:
                    height = rect.get_height()
                    ax.text(rect.get_x()+rect.get_width()*1.0/2., 1.05*height, str(round(height,2)), ha='center', va='bottom')

        autolabel(rects1)
        autolabel(rects2)
        
        plt.show()
```
