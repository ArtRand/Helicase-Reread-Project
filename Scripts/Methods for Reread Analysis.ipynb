## Designing the Hidden Markov Model

<img src = 'pics/Simple.png', width = 500 >

The model started out with a classic sequence alignment HMM structure that handles inserts, deletes, and matches. 

Appended to the structure of that model are the red states which attempt to model the backslip phenomena seen as a characteristic of the helicase motor used.  Each of the backslips has a chance to continue it's precession backwards or transition back to the match state. 

In order to account for the two distinct ways the the system can jump back to a prior state (backslips and helicase dissociations), probabilities were selected such that $\Pr(\text{Length}=1) = \Pr(\text{Length}=6)$. 

If $P_0$ is the starting transition probability from $M$ -> $B_1$, and a transition from $B_i$ to $B_{i+1}$ is $P_i$. Then:

1. Probability of a backslip of length 1 = $\Pr(\text{Length}=1)$ = $P_0 (1 - P_1)$
2. Probability of a backslip greater than 1 = $\Pr(\text{Length} > 1)$ = $P_0 * P_1$


From that you can form the equality: 
\begin{align*}
\frac{\Pr(\text{Length}=i)}{\Pr(\text{Length}>i)} &= \frac{1-P_i}{P_i} 
\end{align*}


Solving for $P_i$:

\begin{align*}
P_i = \frac{(1-P_i)\Pr(\text{Length}>i)}{\Pr(\text{Length}=i)} \\\\
P_i = \frac{\Pr(\text{Length}>i)}{\Pr(\text{Length}>i-1)}
\end{align*}


This allowed the selection of $P_i$'s such that $\Pr(\text{Length}=1) = \Pr(\text{Length}=6)$. A table below shows the transition probabilities and the associated $\Pr(\text{Length}=1)$.

Once the modular board was designed, the complete HMM needed to be constructed.  Profiles for each of the three substrates were curated from control experiments of the three strands.  These profiles were then used to contruct an HMM meant to model a complete Hel308 reread event. 

<img src = 'pics/Full.png', width=1000 >

A complete event was modeled as 31 separate amplitudes, with a state in the HMM being assigned to each amplitude.  The differences in current amplitude between the 3 substrates (Cytosine, Methyl-Cytosine, Hydroxymethyl-Cytosine), were noted across a 5 state window and represented by distinct forks in the model. Each substrate has both its own unique context region and dinucleotide label that are modeled as separate paths.

This window was picked by constructing a heatmap of the profiles and looking at the regions of interest. 

<img src='pics/heatmap_trace.png', width=1000>

The other interesting feature are the meta-transitions that go to a reread silent state that allows certain states in the model the opportunity to transition to the beginning of the event.  This feature allows the model to align to events that have been read more than one time.

## The CHUNK Method

In order to analyze these events, one can use Standard Posterior Decoding to determine which forks the observations were most likely to be emitted from. Unfortunatly, there is significant noise generated by helicase reread events and observations are commonly misaligned to sections of either the context or label region, obfuscating the classification of the event and even causuing misinterpretation of the label which invalidates the entire classification. 

The first step is divide the set of observations into discrete groups that represent likely context or label regions. These 'chunks' are contiguous sets of observations that were likely to have been emitted from either the context or label region. Observations that are not likely to be emitted are used to deliminate the chunks from one another.
<img src='pics/Partition.png' >

A scoring system was constructed in order to associate each chunk with a level of confidence about how likely that series of events was to occur. If we give a generous margin of 5pA for a particular amplitude, then the chance of getting it correct in a uniform distribution of 0-90 is 0.055. For each context, $0.055^5$ = 5.29 x $10^{-7}$.  This is a generous oversimplification as noise tenders to flicker around particular mean with a wide distribution which means noise can match strongly to one more states within a context if they are similar in amplitude. The point is that even if noise is able to match to a couple states it is unlikely that it will match to all of them.

## Method Comparison

Using an untrained HMM, a variety of CHUNK-based methods were implemented to ensure that there wasn't bias inherent to the CHUNK system.

```python
%matplotlib inline

import os

## Store Results
source = '../Data/Results'
trials = []
for root, dirnames, filenames in os.walk(source):
    trials.append(filenames)

profile_trials = trials[1]
trials = trials[0]
    
## Sort Trials:
untrained_fullset = sorted([x for x in trials if 'Untrained' in x and 'Fullset' in x], key=lambda x: x.split('cscore_')[1].split('.txt')[0], reverse=True )
pseudo_fullset = sorted([x for x in trials if 'TEST' in x and 'Profile' in x], key=lambda x: x.split('cscore_')[1].split('.txt')[0], reverse=True )
Trained_no_hmC = sorted([x for x in trials if 'hmC' in x and 'TEST' in x], key=lambda x: x.split('cscore_')[1].split('.txt')[0], reverse=True )
#double_baked = sorted([x for x in trials if 'Double' in x and 'Baked' in x], key=lambda x: x.split('cscore_')[1].split('.txt')[0], reverse=True )
src = '../Data/Results/'
def accuracy_by_filter_score( trial_set, src ):
    
    # Convert Groups of arrays into lists of scores by csore.
    first, last, random = [], [], []
    hmm, best, ind = [], [], []
    sample_sizes = []
    
    for trial_name in trial_set:
            
            trial = np.loadtxt( src + trial_name, delimiter = ',' )
            
            accuracies = trial[::2]
            softcalls = trial[1::2]
            
            f,l,r,h,b,i = accuracies
            first.append( f ); last.append( l ); random.append( r )
            hmm.append( h ) ; best.append( b ); ind.append( i )
         
            sample_sizes.append( trial_name.split('_')[3] )
            
    x = [ 0.9, 0.8, 0.7, 0.6, 0.5, 0.4, 0.3, 0.2, 0.1, 0.0 ]
    
    plt.plot( x, first, label='first' )
    plt.plot( x, last, label='last' ) 
    plt.plot( x, random, label='random' )
    
    plt.plot( x, hmm, label='hmm' )
    plt.plot( x, best, label='best' )
    plt.plot( x, ind, label='ind' )
    
    plt.title( trial_name.split('_')[0]  )
   
    plt.xlabel( 'Chunk Score Cutoff', fontsize=14 )
    plt.ylabel( 'Accuracy', fontsize=14 )
    plt.ylim( [0,1.0] )
    plt.legend(loc='lower left', bbox_to_anchor=(1.02, 0), borderaxespad=0, fontsize=14)
    plt.gca().invert_xaxis()
    plt.show()

accuracy_by_filter_score( untrained_fullset, src )
#accuracy_by_filter_score( pseudo_fullset )
#accuracy_by_filter_score( Trained_no_hmC ) 
#accuracy_by_filter_score( double_baked, src ) 
```
